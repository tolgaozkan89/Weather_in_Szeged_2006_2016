### With function we can quickly check any year, any feature

```python
def grafik(year,feature):
    df_year = df.loc[df["year"]==year]
    sns.relplot(x="month", y=feature, data=df_year, kind="line")
    plt.show()
```

# Feature Engineering

```python
import seaborn as sns
```

```python
corr = df.corr()
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True);
```

```python
top_feature = corr.index[abs(corr["Temperature (C)"]>0.05)]
plt.subplots(figsize=(12,8))
top_corr = df[top_feature].corr()
sns.heatmap(top_corr,annot=True)
plt.show()
```

```python
print("The most important feature")
corr = df.corr()
corr.sort_values(["Temperature (C)"],ascending=False,inplace=True)
corr["Temperature (C)"]
```

# Prediction

```python
import statsmodels.api as sm
```

```python
Y = df["Temperature (C)"]
X = df[["Visibility (km)","month","Wind Bearing (degrees)","Wind Speed (km/h)","year"]]
```

```python
X = sm.add_constant(X)
```

```python
results = sm.OLS(Y,X).fit()
results.summary()
```
